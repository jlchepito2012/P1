```python
import numpy
import matplotlib.pyplot as plt
from scipy import stats

# uniform distribution
x = numpy.random.uniform(0.0, 3.0, 100)
#print(x)
print('Mean %d' %numpy.mean(x))
print('Median %d' %numpy.median(x))
print('Percentile 25 %d' %numpy.percentile(x,25))
print('Percentile 75 %d' %numpy.percentile(x,75))
plt.hist(x, 20)
plt.show()

# normal distribution
y = numpy.random.normal(5.0, 1.0, 100)
#print(y)
print('Mean %d' %numpy.mean(y))
print('Median %d' %numpy.median(y))
print('Percentile 25 %d' %numpy.percentile(y,25))
print('Percentile 75 %d' %numpy.percentile(y,75))
plt.hist(y, 10)
plt.show()

# scatter plot
plt.scatter(x, y)
plt.show()

# lineal regresion
slope, intercept, r, p, std_err = stats.linregress(x, y)
print('R-squares %d' %r)
def myfunc(x):
  return slope * x + intercept
mymodel = list(map(myfunc, x))
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
